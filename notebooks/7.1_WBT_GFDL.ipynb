```python
import xarray as xr
from matplotlib import pyplot as plt
import numpy as np
%load_ext autoreload
%autoreload 2
import pei.laborfunctions as lf
```

# Daily Max WBGT

```python
# Load labor capacity data
ds = xr.open_dataset('../data/processed/GFDL/labor_max_ens1-3.nc').rename({'__xarray_dataarray_variable__':'capacity'})
```

```python
# Population data
ds_pop = xr.open_dataset('../data/processed/GFDL/population_regrid_esm2m.nc')['population']
ds_pop = ds_pop.where(np.isfinite(ds_pop),0)
ds_pop = ds_pop.rename({'latitude':'lat','longitude':'lon'})
```

```python
fig,ax = plt.subplots()
lf.capacity(ds['capacity'],ds_pop,'Global','GFDL',ax)
ax.set_title('Population-Weighted Annual Labor Capacity: Global, RCP 8.5')
```
